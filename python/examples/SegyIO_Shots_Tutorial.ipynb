```python
from SegyIO_Shots import SegyIO_Shots
import numpy as np
import time

from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)
```

```python
segy_file = 'marmousi_2D.segy'

start = time.time()
Marmousi_Shots = SegyIO_Shots(segy_file)
print(f'Shot loaded in {time.time()-start} seconds')
```

```python
Test_Shot = Marmousi_Shots.get_shot(1)

scale = 0.1

plt.show()
plt.imshow(np.transpose(Test_Shot), origin='upper',cmap='Greys', vmin=-scale,vmax=scale)
```
